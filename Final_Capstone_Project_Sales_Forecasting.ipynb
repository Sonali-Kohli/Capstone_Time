# 6. Forecasting using deep learning algorithms:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
```

```python
# Generate synthetic time series data
np.random.seed(42)
date_range = pd.date_range(start='2023-09-01', periods=12, freq='M')
data = np.sin(np.linspace(0, 1 * np.pi, 12)) + np.random.normal(scale=0.5, size=12)

# Create a DataFrame
df = pd.DataFrame({'Date': date_range, 'Value': data})
df.set_index('Date', inplace=True)
```

```python
# Define train and test series
train_size = len(df) - 3
train_series = df.iloc[:train_size]
test_series = df.iloc[train_size:]

import numpy as np
import pandas as pd

# Function to generate synthetic time series data based on existing data
def generate_synthetic_data(df, noise_level=0.5):
    """
    Generate synthetic time series data based on an existing DataFrame.

    Parameters:
    df (pd.DataFrame): The existing time series DataFrame.
    noise_level (float): The standard deviation of the Gaussian noise to be added.

    Returns:
    pd.DataFrame: A DataFrame with synthetic time series data.
    """

    # Generate synthetic data by adding Gaussian noise
    synthetic_values = df['Value'] + np.random.normal(scale=noise_level, size=len(df))

    # Create a new DataFrame with the synthetic data
    synthetic_df = pd.DataFrame({
        'Date': df.index,
        'Synthetic_Value': synthetic_values
    })
    synthetic_df.set_index('Date', inplace=True)

    return synthetic_df

# Example existing DataFrame with time series data
np.random.seed(42)
date_range = pd.date_range(start='2023-01-01', periods=12, freq='M')
actual_values = np.sin(np.linspace(0, 2 * np.pi, 12))
df = pd.DataFrame({'Value': actual_values}, index=date_range)

# Generate synthetic time series data
synthetic_df = generate_synthetic_data(df, noise_level=0.3)

# Print the synthetic DataFrame
print("Synthetic Time Series Data:")
print(synthetic_df)
```

```python
# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(synthetic_df[['Synthetic_Value']])

# Define time step
time_step = 5

# Helper function to create dataset
def create_dataset(series, time_step=1):
    X, y = [], []
    for i in range(len(series) - time_step):
        X.append(series[i:(i + time_step), 0])
        y.append(series[i + time_step, 0])
    return np.array(X), np.array(y)

# Define train size (you can modify this as needed)
train_size = int(len(scaled_data) * 0.8)

# Create train and test datasets
X_train, y_train = create_dataset(scaled_data[:train_size], time_step)
X_test, y_test = create_dataset(scaled_data[train_size - time_step:], time_step)

# Reshape for LSTM [samples, time steps, features]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Print shapes to verify
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(50))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=2, verbose=2)
```

```python
# Make predictions
y3_pred = model.predict(X_test)
y3_pred = scaler.inverse_transform(np.concatenate((np.zeros((y3_pred.shape[0], 1)), y3_pred), axis=1))[:,1]
y3_test_actual = scaler.inverse_transform(np.concatenate((np.zeros((y_test.shape[0], 1)), y_test.reshape(-1, 1)), axis=1))[:,1]

# Evaluate the model
mse3 = mean_squared_error(y3_test_actual, y3_pred)
print(f'Mean Squared Error: {mse3}')

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(df.index[-len(y3_test_actual):], y3_test_actual, label='Actual')
plt.plot(df.index[-len(y3_test_actual):], y3_pred, label='Predicted')
plt.title('LSTM Model Predictions vs Actual Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
